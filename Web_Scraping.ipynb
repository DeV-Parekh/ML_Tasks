```python
import os
import requests
from bs4 import BeautifulSoup
import openpyxl

# Create a directory to save the extracted article text files
if not os.path.exists("/content/articles"):
    os.makedirs("/content/articles")

# Load the input.xlsx file from local drive
input_file_path = "/content/Input.xlsx"
workbook = openpyxl.load_workbook(input_file_path)
sheet = workbook.active
url_info = {}

# Loop through the rows in the sheet
for row in sheet.iter_rows(min_row=2, values_only=True):
    url_id = int(row[0])
    url = row[1]
    response = requests.get(url)
    html_content = response.text
    soup = BeautifulSoup(html_content, "html.parser")
    title = soup.find("h1")
    if title is not None:
        title = title.text
    else:
        title = "Title Not Found"

    text = ""
    content = soup.find_all("p")
    if content is not None:
        for paragraph in content:
            text += paragraph.text + " "
    else:
        text = "Article Text Not Found"
    
    # Save the extracted article title and text in a text file with URL_ID as its file name in the "articles" directory
    with open(f"/content/articles/{url_id}.txt", "w", encoding="utf-8") as file:
        file.write(f"Article Title: {title}\n")
        file.write(f"Article Text: {text}\n")
        url_info[url_id] = url
        print(f'{url_id} done ')

print(url_info)
```

```python
import os
import re

folder_path = '/content/articles'  # Replace with the actual folder path
start_phrase = 'Ranking customer behaviours'
end_phrase = 'Code Review Checklist'

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    # Check if the path is a file
    if os.path.isfile(file_path):  
        with open(file_path, 'r+') as file:
            content = file.read()
            # Use regex to find the start and end phrases and remove text between them
            start_match = re.search(rf"{start_phrase}", content)
            end_match = re.search(rf"{end_phrase}", content)
            if start_match and end_match:
                file.seek(0)
                file.write(content[:start_match.start()] + content[end_match.end():])
                file.truncate()
```

```python
import os
import pandas as pd
from nltk.tokenize import word_tokenize, sent_tokenize
import re

# Function to clean text using stop words list
def clean_text(text, stop_words):
    text = text.lower()  
    text = re.sub(r'[^\w\s]', '', text) 
    tokens = word_tokenize(text)  
    cleaned_tokens = [token for token in tokens if token not in stop_words] 
    cleaned_text = ' '.join(cleaned_tokens)
    return cleaned_text

def count_syllables(word):
    word = word.lower()
    if word.endswith(('es', 'ed')): 
        return max(1, len(re.findall(r'[aeiouy]{1,2}', word[:-2])))
    else:
        return max(1, len(re.findall(r'[aeiouy]{1,2}', word)))

# Load stop words from text files in a given folder
stop_words_folder = '/content/StopWords' 
stop_words = set()
for file in os.listdir(stop_words_folder):
    if file.endswith('.txt'):
        with open(os.path.join(stop_words_folder, file), 'rb') as f:
            binary_data = f.read()
            decoded_data = binary_data.decode('ISO-8859-1') 
            stop_words = set(decoded_data.splitlines())

# Load input text files from the 'articles' directory
input_dir = '/content/articles'  
output_data = []

for file in os.listdir(input_dir):
    if file.endswith('.txt'):
        url_id = int(file.split('.')[0])
        with open(os.path.join(input_dir, file), 'r') as f:
            input_text = f.read()

            # Clean input text
            cleaned_input_text = clean_text(input_text, stop_words)

            # Perform text analysis
            tokens = word_tokenize(cleaned_input_text)
            sentences = sent_tokenize(cleaned_input_text)

            positive_score = sum([1 for token in tokens if token in positive_words])
            negative_score = -1 * sum([-1 for token in tokens if token in negative_words])
            polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)
            subjectivity_score = (positive_score + negative_score) / (len(tokens) + 0.000001)

            avg_sentence_length = len(tokens) / len(sentences)
            complex_word_count = sum([1 for token in tokens if count_syllables(token) > 2])
            percentage_complex_words = complex_word_count / len(tokens)
            fog_index = 0.4 * (avg_sentence_length + percentage_complex_words)

            avg_word_length = sum([len(token) for token in tokens]) / len(tokens)

            personal_pronouns = len(re.findall(r'\b(?:i|we|my|ours|us)\b', input_text.lower()))
            personal_pronouns -= len(re.findall(r'\b(?:us)\b', input_text.lower()))

            word_count = len(tokens)
            syllable_per_word = sum([count_syllables(token) for token in tokens]) / len(tokens)

            # Create a dictionary to store the analysis results
            result = {
                'URL_ID' : url_id,
                'URL' : url_info[url_id],
                'POSITIVE SCORE': positive_score,
                'NEGATIVE SCORE': negative_score,
                'POLARITY SCORE': polarity_score,
                'SUBJECTIVITY SCORE': subjectivity_score,
                'AVERAGE SENTENCE LENGTH': avg_sentence_length,
                'PERCENTAGE OF COMPLEX WORDS': percentage_complex_words,
                'FOG INDEX': fog_index,
                'AVERAGE NO of WORDS PER SENTENCE': avg_sentence_length,
                'COMPLEX WORD COUNT': complex_word_count,
                'WORD COUNT': word_count,
                'SYLLABLES PER WORD': syllable_per_word,
                'PERSONAL PRONOUN COUNT': personal_pronouns,
                'AVERAGE WORD LENGTH': avg_word_length

            }

            # Append the result to the output data list
            output_data.append(result)

            print(f'{file} done')
# Convert the output data list to a dataframe
output_df = pd.DataFrame(output_data)

# Display the output dataframe
print(output_df.head())
```

```python
sorted_df = output_df.sort_values(by='URL_ID',ascending=True)
```
