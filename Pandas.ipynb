# Pandas

Converting list into series of elements

```python
import numpy as np
import pandas as pd
d=[1,2,3,4,5]
pd.Series(data=d)
#convert element lists into series of elements , which have index from 0-5
```

```python
di={'a':1,'b':2,'c':3,'d':4}
pd.Series(di)
#dictionary keys act as index and values with every key act as series values
```

Mathematical Operations on two series

```python
s1=pd.Series([1,2,3,4],['India','China','Nepal','Russia'])
s2=pd.Series([2,3,4,5],['India','China','Bhutan','Russia'])
s3=pd.Series([2,3,4,5],['India','China','Nepal','Russia'])
print('-----------------')
print('s1+s3\n',s1+s3)  #Addition
print('-----------------')
print('s1-s2\n',s1-s2)  #Subtraction
print('-----------------')
print('s1*s2\n',s1*s2)  #Multiplication
print('-----------------')
print('s1/s2\n',s1/s2)  #Division
```

Data Manipulation:Drop Missing Elements


```python
new_df = df2.dropna(axis=0,how='any') # any
new_df
```

```python
neww_df = df2.dropna(axis=0,how='all') # all
neww_df
```

Data Manipulation: Filling Suitable Value


```python
print(df2['temperature'].fillna(value=df2['temperature'].mean())) # Temperature column
print('----------------------')
print(df2['windspeed'].fillna(value=df2['windspeed'].std())) # Windspeed column
```

```python
# Creating a dataframe
df_new = pd.DataFrame(weather_data)
df_new
```

```python
#set column as index
df_new.set_index(['event'],inplace = True)
df_new
```

```python
#Fetch particular rows
df_new.loc['Snow']
```

```python
# Replace function
df = pd.DataFrame({'one':[10,20,30,40,50,2000],'two':[1000,0,30,40,50,60]})
print(df)
print('--------------------')
print(df.replace({1000:10,2000:60})) #1000 replaced with 10, 2000 replaced with 60
```

GroupBy Function

```python
new = df_new.groupby('event').mean() # using mean
print(new)
print('-----------------')
newww = df_new.groupby('event').std() # using std
print(newww)
```

```python
# using groupby to get info for a particular member i.e. Rain
df_new.groupby('event').sum().loc['Rain']
```

```python
# Using max with groupby
df_new.groupby('event').max()
```

```python
# Using unique() and value_counts() to get unique elements and their counts
dff = df_new['windspeed'].unique()
print('-----UNIQUE--------')
print(dff)
print()
print('-----VALUE_COUNTS--------')
dff1 = df_new['windspeed'].value_counts()
print(dff1)
```
